```python
def gridder(X , y , clf,cv):
    from sklearn.model_selection import GridSearchCV   
    param_grid = {
        'iterations': [1200,1400,1600,1700],
        'learning_rate': [0.05],
        'depth': [8,10,12,13],
        'l2_leaf_reg': [3],
        "task_type" : ['GPU']
    }

# Create a GridSearchCV object and fit to the training data
    grid_search = GridSearchCV(clf, param_grid, cv=cv, scoring='f1')
    grid_search.fit(X, y)
    return grid_search.best_params_,grid_search.best_score_
best_param , best_score = gridder(X,y,clf,3)
print(best_score)
print(best_param)
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(iterations=1650 ,depth= 8,l2_leaf_reg= 2,learning_rate= 0.045,border_count=270,bagging_temperature=0.5)
clf.fit(X,y)
```
